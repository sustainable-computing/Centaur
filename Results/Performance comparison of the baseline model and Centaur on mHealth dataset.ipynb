```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd

excel_file = pd.ExcelFile('End-to-end robust multimodal fusion_mHealth.xlsx')
sheet_names = excel_file.sheet_names

sheet_name = 'mode1'  
df_mode1 = excel_file.parse(sheet_name)

# For example, print the first few rows
print(df_mode1.head())
```

```python
cmap = plt.get_cmap("tab10")
my_cmap=['khaki', 'seagreen', 'orange', 'cornflowerblue']
```

```python
#fig, axs = plt.subplots(1, 1,1)


width =0.25

plt.rcParams["figure.figsize"] = "12, 9"
plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"] = "28"

fig, ax = plt.subplots()
ax.grid(linestyle=':',zorder=0)

# plot 1
labels = ['0','0.05','0.1','0.2']

#SADeepSense = [66.17333333,63.47666667,61.32333333,43.47333333] # final
df_centaur = df_mode1[df_mode1["ModelName"] == "Centaur"]
df_units = df_mode1[df_mode1["ModelName"] == "UniTS"]
df_sad = df_mode1[df_mode1["ModelName"] == "SADeepSense"]



x = np.arange(len(labels))  # the label locations
ax.grid(linestyle=':',zorder=0)
rects1 = ax.bar(x - width*0.5, df_centaur["Acc"], width, yerr=df_centaur["Std"], label='Centaur', zorder=3, capsize=10, edgecolor='black',fill=True,color=my_cmap[3], hatch="O")
rects2 = ax.bar(x + width*0.5 , df_units["Acc"], width, yerr= df_units["Std"], label='UniTS', zorder=3, capsize=10, edgecolor='black',fill=True,color=my_cmap[2], hatch="/")
rects3 = ax.bar(x + width*1.5, df_sad["Acc"], width, yerr=df_sad["Std"], label='SADeepSense', zorder=3, capsize=10, edgecolor='black',fill=True,color=my_cmap[1], hatch="X")

hatch="/"
hatch ="-"
ax.set_ylim(0,100)
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy (%)',fontsize=32)
ax.set_xlabel('Noise intensity ()')
ax.set_xlabel('Corruption Mode 1',labelpad = 10, fontsize=32)
# ax.set_title('Scores by group and gender')
ax.set_xticks([0,1,2,3])
ax.set_xticklabels(['$$ =0', '$$ =0.05', '$$= 0.1', '$$= 0.2'])
ax.set_xticks(x, labels)


handles,labels = ax.get_legend_handles_labels()
handles = [handles[0],handles[1], handles[2]]
labels = [labels[0],labels[1], labels[2]]
ax.legend(handles,labels,loc='lower left',fontsize=35,facecolor='white', framealpha=1)
ax.tick_params(axis='x', which='major', pad=15)


fig.savefig('mHealth_corruptionmode1.pdf',  format='eps', dpi=300)
plt.show()
plt.close(fig)
```
